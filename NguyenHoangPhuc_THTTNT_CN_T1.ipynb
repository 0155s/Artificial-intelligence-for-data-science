```python
# Bài 13
def nhap_mang(ten_mang):
    s = input(f"Nhập các phần tử của {ten_mang} (ví dụ: 1 2 4 5): ")
    return [int(x) for x in s.split()]

a = nhap_mang("mảng a")
b = nhap_mang("mảng b")

def tron_mang(a, b):
    tron = []
    for i in range(max(len(a), len(b))):
        val_a = a[i] if i < len(a) else 0
        val_b = b[i] if i < len(b) else 0
        tron.append(val_a + val_b)
    return tron

ket_qua = tron_mang(a, b)
print("\n")
print(f"Mảng a: {a}")
print(f"Mảng b: {b}")
print(f"Mảng kết quả: {ket_qua}")
```

```python
# B14
import numpy as mn

m, n = 4, 5
# a.
a = mn.random.randint(1, 50, size=(m, n))
print("Ma trận a:\n", a)

# b, c.
k = int(input("Nhập k: "))
print(f"Dòng {k} : {a[k, :]}")
print(f"Cột {k} : {a[:, k]}")

# d.
print(f"Dòng có tổng lớn nhất: {mn.argmax(a.sum(axis=1))}")

# f.
print("Phần tử dòng chẵn - cột lẻ:")
print(a[0::2, 1::2])

# h.
mask_bien = mn.ones((m, n), dtype=bool)
mask_bien[1:-1, 1:-1] = False
print(f"TBC biên: {a[mask_bien].mean():.2f}")
# i.
if m > 2 and n > 2:
  ktb =  a[1:-1, 1:-1]
  tich_ktb = ktb.prod()
  n_ktb = ktb.size
  tb_tich = tich_ktb ** (1/n_ktb)
  print(f"Ma trận con: \n{ktb}")
  print(f"Tích các phần tử: {tb_tich:.2f}")
else: print("Phần tử quá nhỏ!")
```

```python
# Bài 15
from datetime import datetime

# a.
class SinhVien:
    def __init__(self, ma, ten, nam_sinh, dtb):
        self.ma = str(ma)[:10]
        self.ten = str(ten)[:20]
        self.nam_sinh = int(nam_sinh)
        self.dtb = float(dtb)

    def __str__(self):
        return f"MSV: {self.ma} | Tên: {self.ten} | Năm sinh: {self.nam_sinh} | DTB: {self.dtb}"

# b.
n = int(input("Nhập số lượng sinh viên: "))
danh_sach = []
for i in range(n):
     print(f"Nhập SV thứ {i+1}:")
     sv = SinhVien(input(" - Mã SV: "), input(" - Tên: "), input(" - Năm sinh: "), input(" - DTB: "))
     danh_sach.append(sv)

print("--- DANH SÁCH SINH VIÊN ---")
for sv in danh_sach: print(sv)
print("-" * 30)

# c.
len_lop = [s for s in danh_sach if s.dtb >= 5]
print(f"c. Số SV đủ điều kiện lên lớp (DTB >= 5): {len(len_lop)}")

# d.
nam_nay = datetime.now().year
print("d. Các sinh viên đủ 20 tuổi:")
for s in danh_sach:
    if nam_nay - s.nam_sinh >= 20:
        print(f"   + {s.ten} ({nam_nay - s.nam_sinh} tuổi)")

# e.
sv_dh = [s for s in danh_sach if s.ma[2:4] == "DH"]
print(f"e. Số SV học hệ Đại học (có 'DH' trong mã): {len(sv_dh)}")

# f.
sv_ten_lan = [s for s in danh_sach if s.ten.split()[-1].lower() == "lan"]
print(f"f. Số sinh viên có tên 'Lan': {len(sv_ten_lan)}")

# g.
sv_ho_phan = [s for s in danh_sach if s.ten.split()[0].lower() == "phan"]
print(f"g. Số sinh viên có họ 'Phan': {len(sv_ho_phan)}")
```
